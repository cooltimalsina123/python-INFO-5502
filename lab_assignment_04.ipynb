```python
# Question 2 -1 (1) Description of all the columns: for each colmun, count number of rows,
# number of unqiue element, the frequency of each unqiue element.
import matplotlib
import pandas as pd

provided_file = pd.read_csv('lab-assignment-04-student-performance.csv')
print("Total columns: ", len(provided_file.columns))

for each_column in provided_file.columns:
    print("Each column name is: ", each_column)
    # we will be using shape[0] to return number of row in column and #value_counts to count unique value in the series
    print("Number of rows in each columns are: ", provided_file[each_column].shape[0])
    print("Number of unique values in each columns are: ", provided_file[each_column].value_counts().shape[0])
    print("Different categories in each column with frequencies of that unique value:")
    print(provided_file[each_column].value_counts())
    print()
```

```python
# (2)  Visualize the count for each value in the 'Study Hours' column. Note:
# Study Hours values comes from this >= 6 hours Good >= 4 hours Average < 2 hours Poor.

import seaborn
import matplotlib.pyplot as plot
%matplotlib inline
print("Frequency for each: ")
print(provided_file['sh'].value_counts())
seaborn.countplot(provided_file['sh'])
plot.title("Study Hours vs Counts")
plot.xlabel("Study Hours")
plot.ylabel("Count")
plot.show()
```

```python

#Question 2-3.
# (3) Using factorplot to show the relationship between numerical and categorical value. Please use figure to show relation between [Study Hours] and [End Semester Percentage]. Percentage is calculated based on the following:
#
# if percentage >=80 then Best
# If percentage >= 60 but less than 80 then Very Good
# If percentage >= 45 but less than 60 then Good
# If Percentage >= 30 but less than 45 then Pass
# If Percentage < 30 then Fail
import pandas as pd
import matplotlib.pyplot as plot
%matplotlib inline
import seaborn
new_data = pd.DataFrame()
new_data['sh']=pd.DataFrame(provided_file.sh)
new_data['esp']=pd.DataFrame(provided_file.esp)
new_data['esp'].mask(new_data['esp'] == 'Best', 80, inplace=True)
new_data['esp'].mask(new_data['esp'] == 'Vg', 60, inplace=True)
new_data['esp'].mask(new_data['esp'] == 'Good', 45, inplace=True)
new_data['esp'].mask(new_data['esp'] == 'Pass', 30, inplace=True)
new_data['esp'].mask(new_data['esp'] == 'Fail', 30, inplace=True)
new_data['exampercent']=pd.DataFrame(provided_file.esp)
seaborn.factorplot(x='sh',y='esp', kind = 'strip',data=new_data)
plot.show()
```

```python
#Question 2-4
# (4) Get the correlation from the data, then plug it to heatmap function and show it as following:
from sklearn import preprocessing
import seaborn
import matplotlib.pyplot as plot
encoding_label = preprocessing.LabelEncoder()
def Encoder(provided_file):
    columnsToEncode = list(provided_file.select_dtypes(include=['category','object']))
    labels = preprocessing.LabelEncoder()

    for f in columnsToEncode:
        provided_file[f] = labels.fit_transform(provided_file[f])
    return provided_file

file_1 = Encoder(provided_file)
file_1 = file_1.drop('ms', axis = 1)
corr = file_1.corr()
seaborn.set(rc = {'figure.figsize':(20,10)})
seaborn.heatmap(corr,annot=True)
plot.show()
```
